###**EXPLORACIÓN DE LOS DATAFRAMES**
####Como tercer paso realizaremos un análisis de la información contenida en nuestros dataframes, para poder comprender así su estructura, el tipo de datos que posee, si hay valores nulos o duplicados, etc.
####Para ello crearemos una función (exploracion_dataframe) que nos permita analizar cada uno de ellos:

```python
def exploracion_dataframe(df):
    print("----------") #Los guiones están con el único objetivo de que al imprimirse la función, los resultados se muestren separados por los mismos
    print("Tamaño del DataFrame")
    print(df.shape) # Con .shape obtengo la cantidad de filas y columnas que posee mi dataframe
    print("----------")
    print("Tipo de datos")
    print(df.info()) # Con .info obtengo los nombres de las columnas y el tipo de datos que poseen, la cantidad de valores NO nulos que hay en cada una, etc.
    print("----------")
    print("Recuento de valores nulos")
    print(df.isnull().sum()) # Obtengo la cantidad de valores nulos que poseo en cada columna
    print("----------")
    print("Recuento de duplicados")
    print(df.duplicated().sum()) # Este método nos proporciona la cantidad de filas duplicadas
    print("----------")
```

####Una vez creada nuestra función, la aplicamos a cada uno de nuestros dataframes:

####Como podemos ver gracias a esta exploración, los dataframes que poseen valores nulos son el de Orders y el de Products. Por ende, vamos a averiguar el porcentaje de nulos en cada uno de ellos:

####*Función porcentaje de nulos:*

```python
def porcentaje_nulos (df):
   porcentaje_nulos = (df.isnull().sum() / len(df)) * 100 #Porcentaje de nulos en cada columna
   porcentaje_nulos = porcentaje_nulos.round(2) #Con .round redondeo decimales, en este caso a 2

   print(porcentaje_nulos)
```

```python
porcentaje_nulos(df_Products) #Aplicamos nuestra función porcentaje_nulos en el dataframe Products
```

###**TRANSFORMACIÓN DE LOS DATAFRAMES**
####Como pudimos ver en nuestra exploración de los dataframes, hay algunos de ellos que poseen valores nulos (los cuales se podrían eliminar en el caso de que su porcentaje no sea muy alto). También hay varias columnas que poseen formato object (por ejemplo, columnas que guardan fechas).
####Para poder realizar una mejor manipulación de nuestra información y evitar problemas a futuro, vamos a pasar las columnas de tipo object a *string* o a formato *fechas* (datetime64[ns]), eliminaremos las filas con valores nulos, y nos aseguraremos de que ciertas columnas estén completamente en *mayúscula* o en *minúscula* (para evitar que una palabra esté escrita con ambos tipos de letra, como por ejemplo Delivered en vez de delivered).

####*Función para transformar datos tipo object a string*:

```python
def object_to_string(df, columnas_string): #Los parametros son el dataframe y la lista con las columnas a transformar
    for columna in columnas_string:
        df[columna] = df[columna].astype("string")
    return df
```

####*Función para transformar datos tipo object a fecha:*

```python
def object_to_date(df, columnas_fechas): #Los parametros son el dataframe y la lista con las columnas a transformar
    df[columnas_fechas] = df[columnas_fechas].apply(pd.to_datetime)
    return df
```

####*Función para convertir minúsculas a mayúsculas:*

```python
def minusc_to_mayusc(df, columna): #Los parametros son el dataframe y la lista con las columna a transformar
    df[columna] = df[columna].str.upper()
    return df
```

####*Función para corroborar si una columna se encuentra escrita en mayúscula  en su totalidad:*

```python
def verificar_mayusculas(df, columna): #Los parámetros son el dataframe y la columna o lista de columnas del mismo que se quiere verificar
  return df[columna].str.isupper().all()
```

###**ANÁLISIS DE LOS DATAFRAMES**
####Ahora realizaremos un análisis de nuestros dataframes ya transformados, para poder sacar la mayor cantidad de datos relevantes de los mismos (*insights*).
####Para llevar esto a cabo, utilizaremos la función *analisisDF*:

```python
def analisisDF(df): #El parámetro de esta función son los dataframes a analizar
    print("--------") #Los guiones están con el único objetivo de que al imprimirse la función, los resultados se muestren separados por los mismos
    print("Analisis de variables cuantitativas")
    columnas_numericas = df.select_dtypes(include=['int', 'float']).columns #Selecciono todas las columnas del dataframe que contienen datos numéricos y obtengo el nombre de estas columnas
    if not columnas_numericas.empty: #Verifico si el df tiene columnas de tipo numéricas y si las hay, se imprime un resumen descriptivo de las mismas con el método .describe()
        print(df[columnas_numericas].describe())
    print("--------")
    print("Analisis de variables cualitativas")
    columnas_string = df.select_dtypes(include=['string']).columns # #Selecciono todas las columnas del df que contienen datos tipo string y obtengo el nombre de estas columnas
    if not columnas_string.empty: #Verifico si el df tiene columnas de tipo string y si las hay, se imprime un resumen descriptivo de las mismas con el método .describe()
        print(df[columnas_string].describe())
```
